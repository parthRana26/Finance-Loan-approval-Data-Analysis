```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Files/Finance Loan approval Data.csv")
```

```python
ctab = pd.crosstab(df['Gender'], df['Married'])
print(ctab)
sns.heatmap(ctab)
```

```python
ctab = pd.crosstab(df['Gender'], df['Education'])
print(ctab)
sns.heatmap(ctab)
```

```python
ctab = pd.crosstab(df['Gender'], df['Self_Employed'])
print(ctab)
sns.heatmap(ctab)
```

```python
ctab = pd.crosstab(df['Gender'], df['Property_Area'])
print(ctab)
sns.heatmap(ctab)
```

```python
plt.figure(figsize=(15, 7))
plt.subplot(2, 2, 1)
sns.barplot(x='Gender', y='ApplicantIncome', data=df, hue='Credit_History')
plt.subplot(2, 2, 2)
sns.barplot(x='Gender', y='ApplicantIncome', data=df, hue='Education')
plt.subplot(2, 2, 3)
sns.barplot(x='Gender', y='LoanAmount', data=df, hue='Married')
plt.subplot(2, 2, 4)
sns.barplot(x='Gender', y='LoanAmount', data=df, hue='Property_Area')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x='Gender', y='LoanAmount', data=df, hue='Property_Area')
plt.subplot(1, 2, 2)
sns.boxplot(x='Gender', y='LoanAmount', data=df, hue='Married')
plt.show()
```

```python
sns.histplot(df[df['Credit_History']==0]['LoanAmount'], kde=True, label='Credit_History=0')
sns.histplot(df[df['Credit_History']==1]['LoanAmount'], kde=True, label='Credit_History=1')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.barplot(x='Gender', y='LoanAmount', data=df, hue='Loan_Amount_Term')
plt.subplot(1, 2, 2)
sns.boxplot(x='Gender', y='LoanAmount', data=df, hue='Credit_History')
plt.show()
```
